```python
import shutil
import os

def move_folder(source_directory, destination_directory, new_name):
    try:
        if not os.path.exists(source_directory):
            print(f"Folder '{source_directory}' tidak ditemukan")
            return

        new_path = os.path.join(destination_directory, new_name)

        if not os.path.exists(destination_directory):
            os.makedirs(destination_directory)

        shutil.move(source_directory, new_path)
        print(f"Folder berhasil dipindahkan dari '{source_directory}' ke '{new_path}'")

    except Exception as e:
        print(f"Terjadi kesalahan: {e}")
```

## **Fungsi Evaluation Metrics**

```python
def confusion_matrix(results):
  a = results.confusion_matrix.matrix[0,0]
  b = results.confusion_matrix.matrix[0,1]
  c = results.confusion_matrix.matrix[0,2]
  d = results.confusion_matrix.matrix[1,0]
  e = results.confusion_matrix.matrix[1,1]
  f = results.confusion_matrix.matrix[1,2]
  g = results.confusion_matrix.matrix[2,0]
  h = results.confusion_matrix.matrix[2,1]
  i = results.confusion_matrix.matrix[2,2]

  TP_x = int(a)
  FP_x = int(d + g)
  FN_x = int(b + c)
  TN_x = int(e + f + h + i)

  TP_y = int(e)
  FP_y = int(b + h)
  FN_y = int(d + f)
  TN_y = int(a + c + g + i)

  print(TP_x, FN_x)
  print(FP_x, TN_x)
  print(TP_y, FN_y)
  print(FP_y, TN_y)

  return TP_x, FP_x, FN_x, TN_x, TP_y, FP_y, FN_y, TN_y
  # Agak laen ini jadi patokan india
  # TP FP
  # FN TN
```

```python
def calculate_metrics(TP, FP, FN, TN):
  # Calculate accuracy, precision, recall, and f1-score
  accuracy = (TP + TN) / (TP + TN + FP + FN) if (TP + TN + FP + FN) != 0 else 0
  precision = TP / (TP + FP) if (TP + FP) != 0 else 0
  recall = TP / (TP + FN) if (TP + FN) != 0 else 0
  f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

  return accuracy, precision, recall, f1_score
```

```python
def calculate_roc_curve(results):
  TP_x, FP_x, FN_x, TN_x, TP_y, FP_y, FN_y, TN_y = confusion_matrix(results)
  # Calculate false positive rate and true positive rate
  # Class Bolt
  fpr_x = FP_x / (FP_x + TN_x) if (FP_x + TN_x) != 0 else 0
  tpr_x = TP_x / (TP_x + FN_x) if (TP_x + FN_x) != 0 else 0

  # Class No_Bolt
  fpr_y = FP_y / (FP_y + TN_y) if (FP_y + TN_y) != 0 else 0
  tpr_y = TP_y / (TP_y + FN_y) if (TP_y + FN_y) != 0 else 0

  print(tpr_x, fpr_x)
  print(tpr_y, fpr_y)

  return fpr_x, tpr_x, fpr_y, tpr_y
```

```python
def calculate_ev_metrics(results):
  # Calculate evaluation metrics
  TP_x, FP_x, FN_x, TN_x, TP_y, FP_y, FN_y, TN_y = confusion_matrix(results)
  #Class Bolt
  accuracy_x, precision_x, recall_x, f1_score_x = calculate_metrics(TP_x, FP_x, FN_x, TN_x)
  #Class No_Bolt
  accuracy_y, precision_y, recall_y, f1_score_y = calculate_metrics(TP_y, FP_y, FN_y, TN_y)

  return (accuracy_x, precision_x, recall_x, f1_score_x, accuracy_y, precision_y, recall_y, f1_score_y)
```

## **Hitung ROC**

```python
def calculate_roc_for_thresholds(model, data_yaml, project_path, iou=0.5, split='test'):
    thresholds = range(10, 101, 10)
    fpr_x_list, tpr_x_list = [], []
    fpr_y_list, tpr_y_list = [], []

    for conf_threshold in thresholds:
        conf = conf_threshold / 100.0
        results = model.val(data=data_yaml, project=project_path, conf=conf, iou=iou, split=split, name=f'conf{conf_threshold}')
        fpr_x, tpr_x, fpr_y, tpr_y = calculate_roc_curve(results)

        fpr_x_list.append(fpr_x)
        tpr_x_list.append(tpr_x)
        fpr_y_list.append(fpr_y)
        tpr_y_list.append(tpr_y)

    roc_data = {
        'Bolt': {'fpr': fpr_x_list, 'tpr': tpr_x_list},
        'No_Bolt': {'fpr': fpr_y_list, 'tpr': tpr_y_list}
    }

    return roc_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc

def plot_roc_curve(roc_data):
    # Calculate AUC for each class
    auc_bolt = auc(roc_data['Bolt']['fpr'], roc_data['Bolt']['tpr'])
    auc_no_bolt = auc(roc_data['No_Bolt']['fpr'], roc_data['No_Bolt']['tpr'])

    # Plot ROC curve for Bolt class
    plt.figure()
    plt.plot(roc_data['Bolt']['fpr'], roc_data['Bolt']['tpr'], color='blue', lw=2,
             label=f'Bolt (AUC = {auc_bolt:.2f})')

    # Plot ROC curve for No_Bolt class
    plt.plot(roc_data['No_Bolt']['fpr'], roc_data['No_Bolt']['tpr'], color='red', lw=2,
             label=f'No_Bolt (AUC = {auc_no_bolt:.2f})')

    plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()
```
